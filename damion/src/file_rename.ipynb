# Rename for Yolov3

## Only 'jpg'

```python
import os

path = os.listdir("C:\\Users\\wnsgk\\Desktop\\data")
count = 1
for i in path:
    name = "C:\\Users\\wnsgk\\Desktop\\data\\" + i
    rename = "C:\\Users\\wnsgk\\Desktop\\data\\kick"+ str(count) + '.jpg'
    count += 1;
    os.rename(name,rename)
    print(rename)
```

## 'jpg' and 'txt'

```python
import os

path = os.listdir("C:\\Users\\wnsgk\\Desktop\\data")
count = 1
for i in path:
    if (i[-3:] =='jpg'):
        image_name = "C:\\Users\\wnsgk\\Desktop\\data\\" + i
        image_rename = "C:\\Users\\wnsgk\\Desktop\\data\\kick"+ str(count) + '.jpg'
        label_name = image_name[:-3] + 'txt'
        label_rename = "C:\\Users\\wnsgk\\Desktop\\data\\kick"+ str(count) + '.txt'
        os.rename(image_name,image_rename)
        os.rename(label_name,label_rename)
        count += 1
        print(image_rename)
        print(label_rename)
```
